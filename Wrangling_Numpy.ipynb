```python
import time

x = [1, 2, 3, 4]
y = [4, 5, 6, 7]
z = []
start = time.time()
for i, j in zip(x, y):
  z.append(i + j)
end = time.time()
print(z)
print((end-start)*10**6,"ns")
```

```python
import numpy as np

x = [1, 2, 3, 4]
y = [4, 5, 6, 7]
start = time.time()
z = np.add(x, y)
end = time.time()
print(z)
print((end-start)*10**6,"ns")
```

```python
import time
start = time.time()
x = [1, 2, 3, 4]
y = [4, 5, 6, 7]
z = []

for i in range(len(x)):
  z.append(x[i]+y[i])
end = time.time()
print((end-start)*10**6,"ns")
print(z)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10* np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=1, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show()
```
